```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import pathlib
import cv2
import pathlib
import torch
import torchvision.transforms as transforms
import cv2

DATA_DIR = "facial_expressions/data"
IMAGE_DIR = "facial_expressions/images"

data = pd.read_csv(f'{DATA_DIR}/legend.csv')
data['emotion'] = data['emotion'].str.lower()
image_paths = list(pathlib.Path(IMAGE_DIR).glob('*.jpg'))
images = np.array([plt.imread(str(path)) for path in image_paths])
```

```python
import torch
df = pd.DataFrame({"path": [p.name for p in image_paths], "image": images})
df.join(data.set_index('image'), on='path')
```
